correlation_coefficient

```python
correlation_coefficient = df["Cases - cumulative total"].corr(df["Deaths - cumulative total"])
correlation_coefficient
```

```python
ax = df.plot.scatter(x="Cases - cumulative total", 
                       y="Deaths - cumulative total", 
                       c='DarkBlue', 
                       loglog=True, 
                       figsize=(14, 8), 
                       title="Relationship between Cumulative Total Cases and Deaths",
                       grid=True, 
                       alpha=0.7)

ax.set_xlabel("Cumulative Total Cases")
ax.set_ylabel("Cumulative Total Deaths")

plt.show()
```
